```python
!pip install pdfplumber
!pip install -U google-generativeai


import pdfplumber
from google.colab import files
import json

# PDFアップロード（A～Fの問題）
# PDFは厚生省が公表しているものを使用する
uploaded = files.upload()
pdf_paths = list(uploaded.keys())
print("アップロードしたPDF一覧:", pdf_paths)

# PDFごとに全ページの文字を抽出してまとめる
all_pdfs_text = []

for pdf_path in pdf_paths:
    pdf_data = {"file_name": pdf_path, "pages": []}
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            text = page.extract_text() or ""
            pdf_data["pages"].append({
                "page_number": page_num + 1,
                "text": text,
                "has_image": bool(page.images)
            })
    all_pdfs_text.append(pdf_data)

# JSON保存
json_path = "all_pdfs_raw_text.json"
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(all_pdfs_text, f, ensure_ascii=False, indent=2)

print("すべてのPDFの文字列をJSONに保存しました:", json_path)
```

```python
import re # 正規表現に関するモジュール

# 問題文を格納する配列
questions = []

# 各pdfファイル（A～Fの問題）に対してそれぞれ処理する
for pdf in all_pdfs_text:
  # 各ページについてそれぞれ処理する
  for page in pdf["pages"]:
    # ページ番号
    page_number=page["page_number"]
    # ページ内のテキスト全文
    text=page["text"]

    if re.match(r"^\d+ ", text):
      questions.append(text)
      print(text)
      print('------------------------------')
```

```python
# Gemini APIをインストール
!pip install -q google-generativeai

import os
import google.generativeai as genai
from dotenv import load_dotenv

# すでにアップロードされている.envを削除する
if os.path.exists(".env"):
  os.remove(".env")

# .envファイル（ローカルにおいておく）をアップロード
# ---------.envの中身---------
# GEMINI_API_KEY=...
# ---------.envの中身---------
uploaded = files.upload()

print("upload完了")

# .envを読み込む
# 引数で上書き読み込みを明示
load_dotenv(override=True)

# 環境変数取得
os.environ["GEMINI_API_KEY"] = os.getenv("GEMINI_API_KEY")

print(os.environ["GEMINI_API_KEY"])
```

```python
# APIキー設定
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# モデル選択
model = genai.GenerativeModel("gemini-2.0-flash")


# --- Geminiに解答依頼 ---
system_prompt = """
あなたは日本の医師国家試験に熟達したAIアシスタントです。
各設問について、病態生理・臨床経過・検査所見を正確に読み取り、
日本の医師国家試験における正答基準に従って最も妥当な選択肢を選んでください。

【出力ルール】
- 出力は必ず以下のJSON配列形式にする：
[
  {"number": 問題番号, "answer": "解答"}
]
- 「推論過程」は出力しない。
- 選択肢は必ず a～e の中から明確に1つ（または設問で指定された個数）選ぶ。
- 不明な場合でも必ず解答を出す。
"""

user_prompt = f"""
以下の問題について、上記ルールに従って解答してください。

【問題文】
{questions}

【思考手順】
1. 問題文を分析し、出題分野（代謝・血液・神経・感染症など）を特定する。
2. 病態の原因と結果を整理して論理的に考える。
3. 臨床的に最も妥当な選択肢を選ぶ。
4. JSON形式で答える。
"""

response = model.generate_content(
    [system_prompt, user_prompt],
    generation_config={"temperature": 1.0},
)

print(response.text)
```
